```python
#import libraries
import pandas as pd
df = pd.read_csv('../data/storedata.csv',encoding='latin')
df.head(20)
```

```python
# rename column names.....make then lower case and replace with underscore
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head(5)
```

```python
#check null values
df.isna().sum()
```

```python
# Convert columns to datetime
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
df['ship_date'] = pd.to_datetime(df['ship_date'], errors='coerce')

# Check how many values became null
print(df[['order_date', 'ship_date']].isna().sum())

# See problematic rows (optional but important)
df[df['order_date'].isna()]
```

```python
df[['sales','profit']]
```

```python
#Find which category + discount combination is causing losses
df.groupby('discount')[['sales','profit']].mean().sort_values(by='profit')
```

```python
df.groupby(['category','discount'])[['sales','profit']].mean().sort_values(by='profit')
```

```python
#which category overall has most losses?
df.groupby('category')['profit'].sum().sort_values()
```

```python
#Which sub-categories in furniture are loss-making?
df[df['category']=='Furniture'].groupby('sub-category')[['sales','profit']].sum().sort_values(by='profit')
```

```python
#create new column
df['profit_margin']=df['profit']/df['sales']
```

```python
#calculate margin 
df.groupby('category')['profit_margin'].mean().sort_values()
```

```python
#find whether high discount is the main reason for losses in tables sub-category?
df[(df['category']=='Furniture') & (df['sub-category']=='Tables')].groupby('discount')[['sales','profit']].mean().sort_values(by='profit')
```

```python
monthly_sales = df.groupby(df['order_date'].dt.to_period('M'))['sales'].sum()
monthly_sales.plot(kind='line', title='Monthly Sales Trend')
```

```python
import matplotlib.pyplot as plt
monthly_sales.plot(kind='line', title='Monthly Sales Trend')
plt.savefig('../outputs/monthly_sales_trend.png')
```

```python
monthly_sales = df.groupby(df['order_date'].dt.to_period('M'))['sales'].sum()

monthly_sales = monthly_sales.to_frame()

monthly_sales['growth_%'] = monthly_sales['sales'].pct_change() * 100

monthly_sales.head(10)
```

```python
monthly_sales.sort_values(by='growth_%', ascending=False).head(5)
```

```python
monthly_sales.sort_values(by='growth_%').head(5)
```

```python
# Business Insights

# 1. Sales consistently decline in January across multiple years, indicating a post-holiday slowdown in demand.
# 2. Sales show strong growth in March across years, suggesting a recurring seasonal spike, possibly due to financial year-end purchasing behavior.
# 3. Overall, sales follow a clear seasonal pattern with a dip in January followed by recovery and peak performance in March.
```

```python
category_sales = df.groupby('category')['sales'].sum().sort_values(ascending=False)
category_sales
```

```python
# Category Insights

# 1. Technology generates the highest sales, indicating strong customer demand and revenue contribution.
# 2. Office Supplies has the lowest sales among all categories, suggesting lower demand or smaller transaction values.
# 3. Furniture performs moderately, but does not outperform Technology, indicating potential for improvement.
```

```python
profit_category = df.groupby('category')['profit'].sum().sort_values(ascending=False)
profit_category
```

```python
# Profit Insights

# 1. Technology generates the highest profit, making it the most valuable category for the business.
# 2. Furniture shows significantly low profit despite having high sales, indicating poor profitability and potential cost or discount issues.
# 3. Office Supplies maintains a balanced performance with moderate sales and strong profit contribution.
```

```python
df.groupby('region')['sales'].sum().sort_values(ascending=False)
```

```python
# Region Insights

# 1. West region generates the highest sales, indicating strong market performance and customer demand.
# 2. South region shows the lowest sales, suggesting weaker market penetration or lower demand.
# 3. East and Central regions perform moderately, indicating stable but improvable performance.
```

```python
# Final Business Insights

# 1. Sales show a strong seasonal pattern with consistent decline in January and peak growth in March.
# 2. Technology is the top-performing category in both sales and profit, making it the key revenue driver.
# 3. Furniture generates high sales but very low profit, indicating poor profitability and potential cost or discount issues.
# 4. West region leads in sales performance, while South region underperforms and needs improvement.
# 5. Overall, the business has strong revenue drivers but faces profitability and regional imbalance challenges.
```
