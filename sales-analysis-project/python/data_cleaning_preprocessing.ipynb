```python
#import libraries
import pandas as pd
df = pd.read_csv('storedata.csv',encoding='latin')
df.head(20)
```

```python
# rename column names.....make then lower case and replace with underscore
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head(5)
```

```python
#check null values
df.isna().sum()
```

```python
# Convert columns to datetime
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
df['ship_date'] = pd.to_datetime(df['ship_date'], errors='coerce')

# Check how many values became null
print(df[['order_date', 'ship_date']].isna().sum())

# See problematic rows (optional but important)
df[df['order_date'].isna()]
```

```python
df[['sales','profit']]
```

```python
#Find which category + discount combination is causing losses
df.groupby('discount')[['sales','profit']].mean().sort_values(by='profit')
```

```python
df.groupby(['category','discount'])[['sales','profit']].mean().sort_values(by='profit')
```

```python
#which category overall has most losses?
df.groupby('category')['profit'].sum().sort_values()
```

```python
#Which sub-categories in furniture are loss-making?
df[df['category']=='Furniture'].groupby('sub-category')[['sales','profit']].sum().sort_values(by='profit')
```

```python
#create new column
df['profit_margin']=df['profit']/df['sales']
```

```python
#calculate margin 
df.groupby('category')['profit_margin'].mean().sort_values()
```

```python
#find whether high discount is the main reason for losses in tables sub-category?
df[(df['category']=='Furniture') & (df['sub-category']=='Tables')].groupby('discount')[['sales','profit']].mean().sort_values(by='profit')
```
